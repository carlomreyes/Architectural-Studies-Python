## Activate Venv

* Venv must be activated per terminal session.

**At directory:**

```python
D:\Education\Coding\Visual Studio Code\Python_\Architecture_Python\Architectural-Studies-Python
```

**Run the following:**

```python
.\.venv_portable\Scripts\Activate.ps1
```

**Prompt should change to:**

```python
(.venv_portable) PS D:\Education\Coding\Visual Studio Code\Python_\Architecture_Python\Architectural-Studies-Python>
```


# Historical Notes

* Virtual environments originate from Unix shells (1990s)
* Windows PowerShell adopted them later
* Windows never auto-binds environments to folders
* Activation is session-based, not project-based

# This design:

* Prevents accidental dependency pollution
* Allows multiple Python versions side-by-side
